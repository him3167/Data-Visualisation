visualization on iris dataset 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
#loading the iris dataset from sklearn liabrary
iris=snr.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.tail(10)
```

```python
#line graph
ax=iris.plot(figsize=(15,8),title='Iris dataset')
ax.set_xlabel('x axis') #total number of rows in the iris dataset
ax.set_ylabel('y axis') #highest magnitude of numerical variables 
```

```python
#drop the species column 
data=iris.drop(['species'],axis=1) #axis=1 means entire column will be deleted from the dataset
```

```python
data.head()
```

```python
#bar graph
data.plot(kind='bar')
```
